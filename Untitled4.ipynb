```python
import cv2
import numpy as np
import pandas as pd
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Define the image size you want to resize to
IMG_SIZE = (64, 64)  # Resize images to 64x64

# Function to read and resize image data
def image_data_getter(X, Y, num_classes):
    X_result = []
    Y_result = []

    for i in X:
        img = cv2.imread(i)
        if img is not None:
            img_resized = cv2.resize(img, IMG_SIZE)  # Resize the image
            img_resized = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB
            X_result.append(img_resized)
        else:
            print(f"Warning: Could not read image at {i}")

    for i in Y:
        Y_result.append(i)

    # One-hot encoding the ordinal labels
    Y_result = to_categorical(Y_result, num_classes=num_classes)
    
    # Normalize the image data (convert pixel values to range [0, 1])
    X_result = np.array(X_result) / 255.0
    
    return np.array(X_result), np.array(Y_result)

# Assuming 'img_paths' and 'labels' are already defined as in your code.
# Create a DataFrame for easier handling
df = pd.DataFrame({
    'IMAGE PATH': img_paths,
    'LABEL': labels
})

# Sort and map classes to numerical labels
classes = sorted(set(labels))  # Sorting ensures consistent class order
label_mapping = {label: idx for idx, label in enumerate(classes)}
df['MAPPED LABELS'] = df['LABEL'].map(label_mapping)

# Split the data into training and test sets
X_data = df['IMAGE PATH']
y_data = df['MAPPED LABELS']
X_train, X_test, Y_train, Y_test = train_test_split(X_data, y_data, shuffle=True, test_size=0.02, stratify=y_data, random_state=42)

# Get training and test image data
X_train_arr, Y_train_arr = image_data_getter(X_train, Y_train, len(classes))
X_test_arr, Y_test_arr = image_data_getter(X_test, Y_test, len(classes))

# Print the dataset sizes
print(f"Total number of Images: {len(X_data)}")
print(f"Number of Training Images: {len(X_train_arr)}")
print(f"Number of Test Images: {len(X_test_arr)}")
print(f"Shape of a single image: {X_train_arr[0].shape}")
```

```python
#Saving the test df so that it can be used later with the saved models
test_df = pd.concat([Y_test, X_test], axis=1)
test_df.columns = ['MAPPED LABELS', 'IMAGE PATH']
test_df.to_csv('anaconda3/test.csv') 
```

```python
from sklearn.utils import class_weight
import numpy as np
from keras import layers, models, regularizers

# 1. Convert Y_train_arr from one-hot encoding back to integer labels (if needed)
Y_train_labels = np.argmax(Y_train_arr, axis=1)  # Convert one-hot encoded to integer labels

# 2. Calculate Class Weights based on the training data
class_weights = class_weight.compute_class_weight(
    'balanced',  # Automatically balances weights based on class frequency
    classes=np.array([0, 1, 2, 3, 4, 5, 6, 7]),  # Class indices for blood groups (0-7)
    y=Y_train_labels  # Use integer labels for class weight calculation
)

# Convert the class weights into a dictionary for the model
class_weight_dict = {i: class_weights[i] for i in range(8)}  # Dictionary for class weights

# 3. Define the model
model = models.Sequential(
    [
        layers.Input(shape=(64, 64, 3)),  # Adjust input shape to match the training data
        layers.Conv2D(32, padding="same", kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        
        layers.Conv2D(32, kernel_size=(3, 3), padding="same", activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        
        layers.Conv2D(64, kernel_size=(3, 3), padding="same", activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        
        layers.Conv2D(64, kernel_size=(3, 3), padding="same", activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        
        layers.Conv2D(128, kernel_size=(3, 3), padding="same", activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        
        layers.Flatten(),  # Flatten the output
        layers.Dropout(0.5),
        
        layers.Dense(128, activation="relu"),  # Adjust first dense layer
        layers.Dense(8, activation="softmax", kernel_regularizer=regularizers.l1_l2(l1=0.01, l2=0.01)),  # Output layer
    ]
)

# Print model summary to verify architecture
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. Train the model with class weights
history = model.fit(x=X_train_arr, y=Y_train_arr, 
                    batch_size=64, 
                    epochs=20, 
                    validation_split=0.2, 
                    class_weight=class_weight_dict)  # Add class_weight here
```

```python
import os
from tensorflow.keras.models import save_model
import pickle

# Define the correct path to the saveModel directory
save_model_path = os.path.join('project', 'saveModel')

# Create the directory if it doesn't exist
if not os.path.exists(save_model_path):
    os.makedirs(save_model_path)

# Save the model with the correct extension (.keras or .h5)
model.save(os.path.join(save_model_path, 'model.keras'))  # Recommended: .keras extension
# Alternatively: model.save(os.path.join(save_model_path, 'model.h5')) for HDF5 format

# Save the training history using pickle
with open(os.path.join(save_model_path, 'model_history.pkl'), 'wb') as history_file:
    pickle.dump(history.history, history_file)

print("Model and history saved successfully!")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
import os

# Load the saved model
model = load_model('project/saveModel/model.keras')  # Change the path if needed

# Define the image size that the model expects
IMG_SIZE = (64, 64)  # Resize images to 64x64 (same as during training)

# Function to load and preprocess the image
def preprocess_image(image_path):
    # Check if the file exists before loading
    if not os.path.isfile(image_path):
        print(f"Error: Image file not found at {image_path}")
        return None
    
    img = cv2.imread(image_path)  # Load the image
    if img is None:
        print(f"Error: Unable to load image at {image_path}")
        return None
    
    img_resized = cv2.resize(img, IMG_SIZE)  # Resize the image to match the model input
    img_resized = img_resized / 255.0  # Normalize the image (same as during training)
    return np.expand_dims(img_resized, axis=0)  # Add batch dimension

# Function to predict the blood group
def predict_blood_group(image_path):
    # Preprocess the image
    img_processed = preprocess_image(image_path)
    if img_processed is None:
        return None, None, None

    # Make prediction
    prediction = model.predict(img_processed)

    # Print the raw output probabilities
    print("Raw Prediction Probabilities: ", prediction)

    # Get the predicted class label
    predicted_class = np.argmax(prediction, axis=1)[0]

    # Map the predicted class to the corresponding blood group
    blood_groups = ["A+", "A-", "AB+", "AB-", "B+", "B-", "O+", "O-"]
    predicted_blood_group = blood_groups[predicted_class]

    return predicted_blood_group, prediction, img_processed

# Example usage: Predict blood group of a sample image
image_path = r"C:\Users\anuna\OneDrive\Desktop\project\saveModel\test\testO+.bmp"  # Corrected path
predicted_blood_group, prediction, img_processed = predict_blood_group(image_path)

# Check if prediction was successful
if predicted_blood_group:
    # Display the image and the predicted blood group
    plt.imshow(img_processed[0])  # Display the image
    plt.title(f"Predicted Blood Group: {predicted_blood_group}")
    plt.axis('off')
    plt.show()

    # Print out the prediction and the predicted blood group
    print(f"Prediction Probabilities: {prediction}")
    print(f"Predicted Blood Group: {predicted_blood_group}")
else:
    print("Prediction failed.")
```
